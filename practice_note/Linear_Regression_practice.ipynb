```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as split
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error as MSE
```

```python
# 資料生成
def datagen():
    
    X = []
    Y = []
    
    for _ in range(30):
        x = 20 * np.random.random() + np.random.randint(-1, 1)
        y = 6 * x - 5 + np.random.random()
        X.append(x)
        Y.append(y)
        
    X = np.array(X).reshape((len(X), 1))
    Y = np.array(Y).reshape((len(Y), 1))
        
    return X, Y

X, Y = datagen()

# 資料分割
train_X, val_X, train_Y, val_Y = split(X, Y, test_size = 0.3)

# 印出數據分佈
plt.scatter(train_X, train_Y, label = 'Train')
plt.scatter(val_X, val_Y, color = 'magenta', marker = '*', label = 'Validation')
plt.legend(loc='upper left')
plt.title('Dataset')
plt.xlabel('rand_X')
plt.ylabel('rand_Y')

plt.plot()
```

```python
#呼叫訓練模型
regs = LR()
regs.fit(train_X, train_Y)
result = regs.predict(val_X)

print('target:\n', val_Y)
print('result:\n', result)
```

```python
#印出訓練預測線 & 驗證學習效果
plt.figure()
plt.scatter(train_X, train_Y, color = 'g')
plt.plot(train_X, regs.predict(train_X), color = 'magenta')
plt.title('Training_Regression')
plt.xlabel('rand_X')
plt.ylabel('rand_Y')

plt.figure()
plt.scatter(val_X, val_Y, color = 'g')
plt.plot(val_X, result, color = 'magenta')
plt.title('Validation_Regression')
plt.xlabel('rand_X')
plt.ylabel('rand_Y')

plt.show()
```

```python
RMSE = MSE(val_Y, result) ** (1/2)
print('RMSE:', RMSE)
```
